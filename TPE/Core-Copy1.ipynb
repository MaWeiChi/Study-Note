# Authentication

#### Sign in

```python
url='https://192.168.3.127/api/v1/auth'
login={
    'name': 'name',
    'password': 'password'
    }
headers={'Content-Type':'application/json'}
```

```python
r=requests.post(url ,json=login,headers=headers)
r.status_code
```

```python
r.json()
```

#### Get token

```python
token=list(r.json().values())[0]['token']
token
```
